```python
# WINTER MONTHS PEAK HOURS
```

```python
winter_df = jn2018.append(nv2018)
winter_df.head()
```

```python
winter_df = winter_df[['tripduration','starttime','start station id','bikeid','gender']]
winter_df.head()
```

```python
# Create an empty list 
# Return time formats
times = []

# Loop through 'starttime' column, split by 'space' then append into our empty list
for x in winter_df.starttime:
    times.append(x.split(' '))
    
# unpack the list 
winter_df['startdate'],winter_df['time'] = zip(*times)

# winter_df.head()
```

```python
# winter_df.dtypes

winter_df.time = winter_df.time.str[0:2]
# winter_df.head()
```

```python
w_time = winter_df.groupby(winter_df.time)
```

```python
updated_winter_times = pd.DataFrame()

updated_winter_times['HourOfDay'] = w_time['time'].first()
updated_winter_times['count'] = w_time['time'].count().unique()

updated_winter_times.head()
```

```python
updated_winter_times.to_csv('2018winter_hours.csv')
```

```python
#TOP AND BOTTOM 10 STATIONS IN NOVEMBER
```

```python
nv2018 = nv2018[['tripduration', 'starttime', 'start station id', 'start station name',
                       'end station id', 'end station name',
                           'bikeid', 'gender']]
nv2018.head()
```

```python
# unpack time

nov2018 = []

# Loop through 'starttime' column, split by 'space' then append into our empty list
for x in nv2018.starttime:
    nov2018.append(x.split(' '))
    
# create 2 columns and unzip the list 
nv2018['startdate'],nv2018['time'] = zip(*nov2018)

nv2018.head()
```

```python
station_id = nv2018.groupby(nv2018['start station id'])

strtstation_id = pd.DataFrame()

strtstation_id['station_id'] = station_id['start station id'].first()
strtstation_id['count'] = station_id['start station id'].count()
strtstation_id['station_name'] = station_id['start station name'].first()
```

```python
top_start = strtstation_id.sort_values(by='count', ascending=False).head(10)
# top_start
top_start.to_csv('top10_start_nov2018.csv')
```

```python
bottom_start = strtstation_id.sort_values(by='count', ascending=False).head(10)
# bottom_start
bottom_start.to_csv('bottom10_start_nov2018.csv')
```

```python
# END STATIONS (TOP AND BOTTOM FOR NOVEMBER)
end_station = nv2018.groupby(nv2018['end station id'])

end_station_id = pd.DataFrame()

end_station_id['station_id'] = end_station['end station id'].first()
end_station_id['count'] = end_station['end station id'].count()
end_station_id['station_name'] = end_station['end station name'].first()
```

```python
top_end = end_station_id.sort_values(by='count', ascending=False).head(10)
# top_end
top_end.to_csv('top10_end_nov2018.csv')
```

```python
bottom_end = end_station_id.sort_values(by='count', ascending=True).head(10)
# bottom_end
bottom_end.to_csv('bottom10_end_nov2018.csv')
```
